##### Ejercicio 6
Selecciona todas las filas donde la columna "CIUDAD" es igual a "Tokio" y la columna "EDAD" es mayor a 45.

##### Ejercicio 7
Agrupa el conjunto de datos por la columna "GENERO" y utiliza el mean y median para obtener la edad promedio y mediana de cada grupo.

```python
dfgroup_A = datos.groupby(['GENERO']).agg({
    'EDAD': ['mean', 'median']})
dfgroup_A
```

##### Ejercicio 8
Utiliza el método pivot_table() para crear una tabla dinámica que muestre la cantidad de personas por GENERO y CIUDAD.

```python
df_pivot = datos[['GENERO','CIUDAD','ID']].pivot_table(index='GENERO',columns=['CIUDAD'],aggfunc='sum')
df_pivot
```

##### Ejercicio 9
Utiliza el método groupby() para agrupar el conjunto de datos por la columna "CIUDAD" y utiliza el método sum() para obtener la suma de las edades de cada grupo.

```python
datos.groupby(['CIUDAD'])[['EDAD']].sum()
```

##### Ejercicio 10
Utiliza el método merge() para unir el conjunto de datos "datos" con otro conjunto de datos llamado informacion_adicional en base a la columna "ID".

```python
pd.merge(datos,informacion_adicional,on='ID')
```

```python
#Aquí agrega las nuevas columnas, pero en las filas donde no hay más valores ID del conjunto de datos informacion_adicional le agrega valores nulos.
pd.merge(datos,informacion_adicional,how='outer',on='ID')
```

## 2. Curiosidades
¿Cómo generar una base de datos con valores aleatorios?

```python
#!pip install faker
```

```python
from faker import Faker
```

```python
# Crear un generador de datos ficticios
fake = Faker()

# Configurar la semilla aleatoria para reproducibilidad
np.random.seed(42)

# Cantidad de filas
n_filas = 1000

# Generar datos aleatorios (creamos un diccionario)
data = {
    'ID': range(1, n_filas+1),
    'NOMBRE': [fake.name() for _ in range(n_filas)],
    'EDAD': np.random.randint(18, 80, size=n_filas),
    'GENERO': np.random.choice(['Masculino', 'Femenino'], size=n_filas),
    'CIUDAD': [np.random.choice(['Nueva York', 'Los Ángeles', 'Londres', 'París', 'Tokio']) for _ in range(n_filas)]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
df.to_csv('datos.csv',sep=";",index=False)
```
